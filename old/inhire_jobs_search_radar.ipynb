# InHire Jobs Radar via Search Engine

Busca vagas em páginas InHire usando DuckDuckGo.

```python
import requests
import pandas as pd
import time
import json
```

```python
COMPANIES = [
"sympla",
#"agenciacriativa","agrosearch","alice","alun","amcom","auvotecnologia",
#"bankme","betaonline","brasilparalelo","ceisc","cerc","cielo","cora",
#"crown","db1","deloitte","digix","fitenergia","flash","flutterbrazil",
#"fretadao","gex","grupoescalar","gx2","infotecbrasil","kanastra","kpmg",
#"livemode","magalu","magazord","milvus","mjv","nomadglobal","olist",
#"openlabs","orizon","paytrack","pier","premiersoft","qive","radix",
#"residclub","sanar","sharepeoplehub","shareprime","supertex","sylvamo",
#"talentt","talentx","tripla","unimar","upflux","v360",
#"v4company","vitru","warren","zig","contabilizei","kiwify","loft",
#"nubank","creditas","ifood","stone","loggi"
]
```

## Palavras-chave

```python
KEYWORDS = [
"analista de dados",
"data analyst",
"data scientist",
"data engineer",
"analytics",
"bi"
]
```

## Função de busca DuckDuckGo

```python
def search_jobs(company, keyword):

    query = f"site:{company}.inhire.app/vagas {keyword}"

    url = "https://duckduckgo.com/html/"

    params = {
        "q": query
    }

    try:

        r = requests.get(url, params=params, headers={"User-Agent":"Mozilla/5.0"})

        html = r.text

        links = []

        for line in html.split('href="'):

            if "inhire.app/vagas" in line:

                link = line.split('"')[0]

                if "http" in link:

                    links.append(link)

        return list(set(links))

    except:

        return []
```

## Scanner

```python
all_jobs = []

for company in COMPANIES:

    print("Escaneando:", company)

    for k in KEYWORDS:

        links = search_jobs(company, k)

        for link in links:

            all_jobs.append({
                "empresa": company,
                "keyword": k,
                "link": link
            })

        time.sleep(1)
```

## Remover duplicados

```python
df = pd.DataFrame(all_jobs)

df = df.drop_duplicates(subset=["link"])

print("Total vagas encontradas:", len(df))

df.head()
```

## Salvar CSV

```python
df.to_csv("vagas_inhire_data_jobs.csv", index=False)

print("Arquivo salvo: vagas_inhire_data_jobs.csv")
```
